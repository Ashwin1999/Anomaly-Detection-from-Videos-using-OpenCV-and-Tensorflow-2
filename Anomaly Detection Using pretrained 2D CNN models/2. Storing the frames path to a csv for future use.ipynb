```python
import os

crimePath, normalPath = "dataset/ImageDataset/crime/", "dataset/ImageDataset/normal/"

crimeFrames, normalFrames = os.listdir(crimePath), os.listdir(normalPath)

crimeFrames = [crimePath+frame for frame in crimeFrames]
normalFrames = [normalPath+frame for frame in normalFrames]

crimeClass = ['crime']*len(crimeFrames)
normalClass = ['normal']*len(normalFrames)

allFrames = crimeFrames+normalFrames
allClass = crimeClass+normalClass
```

```python
import pandas as pd

df = pd.DataFrame({
    'path':allFrames,
    'class': allClass
})

df.head()
```

```python
test = df.sample(frac=1)
test.reset_index(inplace=True)
test.drop('index', axis=1, inplace=True)
test.to_csv('train.csv', index=False)
test
```
