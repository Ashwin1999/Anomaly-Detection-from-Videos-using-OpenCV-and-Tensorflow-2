```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('filePath.csv')
df.head()
```

```python
crime = df[df['class']=='crime']
normal = df[df['class']=='normal']

print(len(crime), len(normal))
```

```python
crimeFramesPath = 'dataset/ImageDataset/crime/'
normalFramesPath = 'dataset/ImageDataset/normal/'
```

```python
crimeVideos = crime.path
normalVideos = normal.path
```

```python
import cv2
```

```python
for video in crimeVideos:
    name = video.split('/')[-1].split('.')[0]
    print(name)
    
    cap = cv2.VideoCapture(video)
    i, n = 1, 30
    print(f'preparing to extract frames from {name}.mp4')
    while cap.isOpened():
        ret, frame = cap.read()
        if ret == False:
            break
        if i%n==0:
            cv2.imwrite(crimeFramesPath+name+str(i/n)+'.jpg',frame)
            i+=1
        else:
            i+=1
    print(f'done extracting from {name}.mp4')
```

```python
for video in normalVideos:
    name = video.split('/')[-1].split('.')[0]
    print(name)
    
    cap = cv2.VideoCapture(video)
    i, n = 1, 30
    print(f'preparing to extract frames from {name}.mp4')
    while cap.isOpened():
        ret, frame = cap.read()
        if ret == False:
            break
        if i%n==0:
            cv2.imwrite(normalFramesPath+name+str(i/n)+'.jpg',frame)
            i+=1
        else:
            i+=1
    print(f'done extracting from {name}.mp4')
```
