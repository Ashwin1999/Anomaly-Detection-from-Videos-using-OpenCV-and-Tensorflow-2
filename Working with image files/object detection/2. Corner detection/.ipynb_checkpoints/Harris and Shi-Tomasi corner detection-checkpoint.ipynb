```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_src = 'Chess.webp'
```

# Harris corner detection

```python
img = cv2.imread(img_src)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray_board = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_board, cmap='gray')
```

```python
dest = cv2.cornerHarris(gray_board, blockSize=2,ksize=3,k=0.04)
# optionally we can give dialate, for better visual.
dest = cv2.dilate(dest,None)
```

```python
img[dest>dest.max()*0.01] = [0,255,0]
plt.imshow(img)

# the markings are there but very faint because of the image contrast
```

# Shi-Tomasi Detection

```python
img1 = cv2.imread(img_src)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread(img_src)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
gray_board1 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_board, cmap='gray')
```

```python
gray_board2 = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_board2, cmap='gray')
```

```python
maxCor = 10
corners = cv2.goodFeaturesToTrack(gray_board1, maxCor, 0.05, 10)

for corner in corners:
    # unpack the corners
    x,y = corner.ravel()
    cv2.circle(img1, (x,y), 3, (0,255,0), 5)
    
plt.imshow(img1)
```

```python
maxCor = 80
corners = cv2.goodFeaturesToTrack(gray_board2, maxCor, 0.05, 10)

for corner in corners:
    # unpack the corners
    x,y = corner.ravel()
    cv2.circle(img2, (x,y), 3, (0,255,0), 5)
    
plt.imshow(img2)
```
