```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('puppy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
edge = cv2.Canny(img, 0, 255)
plt.imshow(edge)
```

```python
x = 2
blured = cv2.blur(img, (x,x))
plt.imshow(blured)
```

```python
med = np.median(blured)

# set the lower and higher threshold to the below values
lower = int(max(0, 0.7*med))
higher = int(max(255, 1.3*med))

# perform edge detectionss
edge = cv2.Canny(blured, lower, higher)
plt.imshow(edge)
```
