```python
import cv2
import matplotlib.pyplot as plt
```

```python
# lets call our puppy

puppy = cv2.imread('puppy.jpg')
```

```python
# lets check what type of puppy we've got

type(puppy)
```

```python
# lets check the shape of our puppy

puppy.shape
```

```python
# let us see the  puppy

plt.imshow(puppy)
```

```python
# hmm...is that puppy a mutant or what?! blue paws and face!
# cv2 reads the color in order BGR. But matplotlib expects color in the order RGB.
# So thats why the puppy is having a shade of blue on its paws and face

# to convert it to RGB we do the following
converter = cv2.COLOR_BGR2RGB
puppy = cv2.cvtColor(puppy,converter)

plt.imshow(puppy)
```

```python
# lets try converting to grayscale
puppy_gray = cv2.cvtColor(puppy,cv2.COLOR_BGR2GRAY)

plt.imshow(puppy_gray)
```

```python
# what?! it looks green!!!
# lets try that again shall we

plt.imshow(puppy_gray, cmap="gray")

# much better!!!
```

```python
# lets make our puppy large
width, height = 5000, 4000
dims = (width, height)

large_puppy=cv2.resize(puppy, dims)
plt.imshow(large_puppy, cmap="gray")
```

```python
# the image size changed to 4000, 5000! yay!!!
large_puppy.shape
```

```python
# lets make the puppy half the dimension of the above image
width_ratio, height_ratio = 0.5, 0.5

# the way we pass the params are a bit wierd
large_puppy=cv2.resize(large_puppy, (0,0), large_puppy, width_ratio, height_ratio)
plt.imshow(large_puppy, cmap="gray")
```

```python
# doble check our result by checking the size
large_puppy.shape
```

```python
# vertical flip
flip1= cv2.flip(large_puppy, 0)
plt.imshow(flip1)

# Shit!!! The puppy will get scared! Lets put it back to normal orientation
```

```python
flip1 = cv2.flip(flip1, 0)
plt.imshow(flip1)
# Phew! thats better!
```

```python
# flip the puppy horizontally
flip2 = cv2.flip(flip1, 1)
plt.imshow(flip2)
# Awww! that is cute!
```

```python
# Thats perfect! Lets save this image!
plt.imsave('cute-puppy.jpg', flip2)
# check your folder if its saved or...
```

```python
check_saved = cv2.imread('cute-puppy.jpg')
check_saved = cv2.cvtColor(check_saved,cv2.COLOR_BGR2RGB)
plt.imshow(check_saved)
# We see it is saved since we're able to imread it
```
